```python
pbp_df = pd.read_csv('../data/raw/basketball.pbps.csv')

pbp_df = pbp_df.drop(['oPlayer.playerId','dPlayer.playerId','prPlayer.playerId','offensiveLineup.0.firstName',
                      'offensiveLineup.0.lastName','offensiveLineup.0.playerId','offensiveLineup.1.firstName',
                      'offensiveLineup.1.lastName','offensiveLineup.1.playerId','offensiveLineup.2.firstName',
                      'offensiveLineup.2.lastName','offensiveLineup.2.playerId','offensiveLineup.3.firstName',
                      'offensiveLineup.3.lastName','offensiveLineup.3.playerId','offensiveLineup.4.firstName',
                      'offensiveLineup.4.lastName','offensiveLineup.4.playerId','defensiveLineup.0.firstName',
                      'defensiveLineup.0.lastName','defensiveLineup.0.playerId','defensiveLineup.1.firstName',
                      'defensiveLineup.1.lastName','defensiveLineup.1.playerId','defensiveLineup.2.firstName',
                      'defensiveLineup.2.lastName','defensiveLineup.2.playerId','defensiveLineup.3.firstName',
                      'defensiveLineup.3.lastName','defensiveLineup.3.playerId','defensiveLineup.4.firstName',
                      'defensiveLineup.4.lastName','defensiveLineup.4.playerId'],axis=1)

pbp_df['Off_Lineup'] = ''
pbp_df['Def_Lineup'] = ''
pbp_df['Shot'] = ''
pbp_df['Shooter'] = ''

pbp_df['TOV'] = ''
pbp_df['TOV_Player'] = ''

pbp_df['PF'] = ''
pbp_df['PF_Player'] = ''

pbp_df['DREB'] = ''
pbp_df['OREB'] = ''
pbp_df['AST'] = ''
pbp_df['STL'] = ''
pbp_df['BLK'] = ''

pbp_df['DREB_Player'] = ''
pbp_df['OREB_Player'] = ''
pbp_df['AST_Player'] = ''
pbp_df['STL_Player'] = ''
pbp_df['BLK_Player'] = ''

pbp_df['FTA'] = ''

pbp_df['FT_Made'] = ''
pbp_df['FT_Miss'] = ''

pbp_df['3PA'] = ''

pbp_df['3P_Made'] = ''
pbp_df['3P_Miss'] = ''

pbp_df['2PA'] = ''

pbp_df['2P_Made'] = ''
pbp_df['2P_Miss'] = ''

pbp_df['Open_Shot'] = ''
pbp_df['High P&R'] = ''
pbp_df['ISO'] = ''
pbp_df['Transition'] = ''
pbp_df['Post-Up'] = ''
pbp_df['Spot-Up'] = ''
pbp_df['Cut'] = ''
pbp_df['Flash'] = ''

for idx,row in pbp_df.iterrows():
    my_list = []
    
    my_list.append(row['offensiveLineup.0.name'])
    my_list.append(row['offensiveLineup.1.name'])
    my_list.append(row['offensiveLineup.2.name'])
    my_list.append(row['offensiveLineup.3.name'])
    my_list.append(row['offensiveLineup.4.name'])
    
    my_list = sorted(my_list)
    
    pbp_df.at[idx,'Off_Lineup'] = str(my_list)
    
    my_list = []
    
    my_list.append(row['defensiveLineup.0.name'])
    my_list.append(row['defensiveLineup.1.name'])
    my_list.append(row['defensiveLineup.2.name'])
    my_list.append(row['defensiveLineup.3.name'])
    my_list.append(row['defensiveLineup.4.name'])
    
    my_list = sorted(my_list)
    
    pbp_df.at[idx,'Def_Lineup'] = str(my_list)
    
    ############################

    my_list = row['secondaryString'].split('>')
    final_my_list = []
    
    for item in my_list:
        final_my_list.append(item.strip())
    
    if final_my_list[0] == 'Shot':
        pbp_df.at[idx,'Shot'] = 1
        pbp_df.at[idx,'Shooter'] = final_my_list[1]
        
        if final_my_list[2] == 'Any Type':
            if '2' in final_my_list[3]:# 2 pointer
                pbp_df.at[idx,'2PA'] = 1
                
                if 'Make' in final_my_list[4]:
                    pbp_df.at[idx,'2P_Made'] = 1
                else:
                    pbp_df.at[idx,'2P_Miss'] = 1
                
            else: # 3 pointer
                pbp_df.at[idx,'3PA'] = 1
                
                if 'Make' in final_my_list[4]:
                    pbp_df.at[idx,'3P_Made'] = 1
                else:
                    pbp_df.at[idx,'3P_Miss'] = 1
            
        else: # FREE THROWS
            pbp_df.at[idx,'FTA'] = 1
            
            if final_my_list[3] == 'Made':
                pbp_df.at[idx,'FT_Made'] = 1
            else:
                pbp_df.at[idx,'FT_Miss'] = 1
        
    elif final_my_list[0] == 'Turnover':
        pbp_df.at[idx,'TOV'] = 1
        pbp_df.at[idx,'TOV_Player'] = final_my_list[2]
        
    elif final_my_list[0] == 'Personal Foul':
        pbp_df.at[idx,'PF'] = 1
        pbp_df.at[idx,'PF_Player'] = final_my_list[2]
        
    elif final_my_list[0] == 'Mist Stat':
        if final_my_list[1] == 'Defensive Rebound':
            pbp_df.at[idx,'DREB'] = 1
            pbp_df.at[idx,'DREB_Player'] = final_my_list[2]
            
        elif final_my_list[1] == 'Offensive Rebound':
            pbp_df.at[idx,'OREB'] = 1
            pbp_df.at[idx,'OREB_Player'] = final_my_list[2]
            
        elif final_my_list[1] == 'Assist':
            pbp_df.at[idx,'AST'] = 1
            pbp_df.at[idx,'AST_Player'] = final_my_list[2]
            
        elif final_my_list[1] == 'Steal':
            pbp_df.at[idx,'STL'] = 1
            pbp_df.at[idx,'STL_Player'] = final_my_list[2]
            
        elif final_my_list[1] == 'Block':
            pbp_df.at[idx,'BLK'] = 1
            pbp_df.at[idx,'BLK_Player'] = final_my_list[2]
    
    ###############################
    if 'Open' in row['offensiveString']:
        pbp_df.at[idx,'Open_Shot'] = True
    
    if 'High P&R' in row['offensiveString']:
        pbp_df.at[idx,'High P&R'] = True
        
    if 'ISO' in row['offensiveString']:
        pbp_df.at[idx,'ISO'] = True
        
    if 'Transition' in row['offensiveString']:
        pbp_df.at[idx,'Transition'] = True
    
    if 'Post-Up' in row['offensiveString']:
        pbp_df.at[idx,'Post-Up'] = True
        
    if 'Spot-Up' in row['offensiveString']:
        pbp_df.at[idx,'Spot-Up'] = True
        
    if 'Cut' in row['offensiveString']:
        pbp_df.at[idx,'Cut'] = True
    
    if 'Flash' in row['offensiveString']:
        pbp_df.at[idx,'Flash'] = True
```

```python
pbp_df = pbp_df[~pbp_df['offensiveLineup.0.name'].isna()]

pbp_df = pbp_df.reset_index()

try:
    pbp_df = pbp_df.drop('index',axis=1)
except:
    pass

pbp_df = pbp_df[['playByPlayId',
                 'game.awayTeam.fullName',
                 'game.awayTeam.awayTeamId',
                 'game.homeTeam.fullName',
                 'game.homeTeam.homeTeamId',
                 'game.gameId',
                 'game.date',
                 'clock',
                 'ato',
                 'eob',
                 'shot',
                 'sob',
                 'zone',
                 'possessionId',
                 'oPlayer.name',
                 'dPlayer.name',
                 'prPlayer.name',
                 'gameHalf',
                 'shotClock',
                 'homeEvent',
                 'turnover',
                 'offensiveString',
                 'secondaryString',
                 'ftAttempt',
                 'ftMade',
                 'Off_Lineup',
                 'Def_Lineup',
                 'Shot',
                 'Shooter',
                 'TOV',
                 'TOV_Player',
                 'PF',
                 'PF_Player',
                 'DREB',
                 'OREB',
                 'AST',
                 'STL',
                 'BLK',
                 'DREB_Player',
                 'OREB_Player',
                 'AST_Player',
                 'STL_Player',
                 'BLK_Player',
                 'FTA',
                 'FT_Made',
                 'FT_Miss',
                 '3PA',
                 '3P_Made',
                 '3P_Miss',
                 '2PA',
                 '2P_Made',
                 '2P_Miss',
                 'Open_Shot',
                 'High P&R',
                 'ISO',
                 'Transition',
                 'Post-Up',
                 'Spot-Up',
                 'Cut',
                 'Flash']]

pbp_df.to_csv('../data/interim/cleaned_pbp_data.csv')
```
