```python
import pandas as pd
import numpy as np
import missingno as msno

pd.options.display.max_columns = 500
pd.options.display.max_rows = 500

df = pd.read_csv('../data/interim/cleaned_pbp_data.csv')
try:
    df = df.drop('Unnamed: 0',axis=1)
except:
    pass

df['seconds_left_in_game'] = 0

# Time Left in Game
final_df = pd.DataFrame()

game_id_list = df['game.gameId']
game_id_list = list(set(game_id_list))

for game in game_id_list:
    cut_df = df[df['game.gameId']==game]
    
    max_period = cut_df['gameHalf'].max()
    
    if max_period == 2:
        for idx,row in cut_df.iterrows():
            if row['gameHalf'] == 2:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10
            else:
                cut_df.at[idx,'seconds_left_in_game'] = (int(row['clock'])/10) + 1200
                
    elif max_period == 3:
        for idx,row in cut_df.iterrows():
            if row['gameHalf'] == 3:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10
            elif row['gameHalf'] == 2:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 300
            else:
                cut_df.at[idx,'seconds_left_in_game'] = (int(row['clock'])/10) + 1200 + 300
    elif max_period == 4:
        for idx,row in cut_df.iterrows():
            if row['gameHalf'] == 4:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10
            elif row['gameHalf'] == 3:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 300
            elif row['gameHalf'] == 2:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 600
            else:
                cut_df.at[idx,'seconds_left_in_game'] = (int(row['clock'])/10) + 1200 + 600
    elif max_period == 5:
        for idx,row in cut_df.iterrows():
            if row['gameHalf'] == 4:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10
            elif row['gameHalf'] == 4:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 300
            elif row['gameHalf'] == 3:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 600
            elif row['gameHalf'] == 2:
                cut_df.at[idx,'seconds_left_in_game'] = int(row['clock'])/10 + 900
            else:
                cut_df.at[idx,'seconds_left_in_game'] = (int(row['clock'])/10) + 1200 + 900
    
    final_df = pd.concat([final_df,cut_df])
    

final_final_df = pd.DataFrame()

for game in game_id_list:
    cut_df = final_df[final_df['game.gameId']==game]
    
    max_time = cut_df['seconds_left_in_game'].max()
    
    cut_df['seconds_into_game'] = max_time - cut_df['seconds_left_in_game']
    
    final_final_df = pd.concat([final_final_df,cut_df])
    
final_final_df.info()
```

```python
final_final_df.to_csv('../data/interim/cleaned_pbp_data.csv')
```

```python
# Running Score
import pandas as pd
import numpy as np
import missingno as msno

pd.options.display.max_columns = 500
pd.options.display.max_rows = 500

df = pd.read_csv('../data/interim/cleaned_pbp_data.csv')
try:
    df = df.drop('Unnamed: 0',axis=1)
except:
    pass

possessions_df = pd.read_csv('../data/raw/basketball possessions.csv')

possessions_df = possessions_df[['defense.teamId','defense.name','offense.teamId','offense.name','possessionId']]

df = df.merge(possessions_df,how = 'left',on = 'possessionId')
```

```python
df = df[['playByPlayId','game.awayTeam.fullName','game.awayTeam.awayTeamId','game.homeTeam.fullName','game.homeTeam.homeTeamId','defense.teamId',
         'defense.name','offense.teamId','offense.name','game.gameId','game.date','clock','ato','eob','shot','sob','zone','possessionId',
         'oPlayer.name','dPlayer.name','prPlayer.name','gameHalf','shotClock','homeEvent','turnover','offensiveString','secondaryString','ftAttempt',
         'ftMade','Off_Lineup','Def_Lineup','Shot','Shooter','TOV','TOV_Player','PF','PF_Player','DREB','OREB',
         'AST','STL','BLK','DREB_Player','OREB_Player','AST_Player','STL_Player','BLK_Player','FTA','FT_Made','FT_Miss','3PA','3P_Made',
         '3P_Miss','2PA','2P_Made','2P_Miss','Open_Shot','High P&R','ISO','Transition','Post-Up','Spot-Up','Cut','Flash','seconds_left_in_game',
         'seconds_into_game']]

df['Home_Team_LiveScore'] = 0
df['Away_Team_LiveScore'] = 0

final_df = pd.DataFrame()

game_id_list = df['game.gameId']
game_id_list = list(set(game_id_list))

for game in game_id_list:
    cut_df = df[df['game.gameId']==game]
    
    cut_df = cut_df.sort_values(by='seconds_left_in_game',ascending=False)
    cut_df = cut_df.reset_index()
    try:
        cut_df = cut_df.drop('index',axis=1)
        
    except:
        pass
    
    my_home_score = 0
    my_away_score = 0
    
    for idx,row in cut_df.iterrows():
        if row['offense.teamId'] == row['game.homeTeam.homeTeamId']:
            if row['FT_Made'] == 1:
                my_home_score = my_home_score + 1
            elif row['3P_Made'] == 1:
                my_home_score = my_home_score + 3
            elif row['2P_Made'] == 1:
                my_home_score = my_home_score + 2
        else:
            if row['FT_Made'] == 1:
                my_away_score = my_away_score + 1
            elif row['3P_Made'] == 1:
                my_away_score = my_away_score + 3
            elif row['2P_Made'] == 1:
                my_away_score = my_away_score + 2
            
        cut_df.at[idx,'Home_Team_LiveScore'] = my_home_score
        cut_df.at[idx,'Away_Team_LiveScore'] = my_away_score
        
    final_df = pd.concat([final_df,cut_df])
    
final_df = final_df.reset_index()
final_df = final_df.drop('index',axis=1)
```

```python
final_df.to_csv('../data/interim/cleaned_pbp_data.csv')
```

```python
# Possession Value
import pandas as pd
import numpy as np
import missingno as msno

pd.options.display.max_columns = 500
pd.options.display.max_rows = 500

df = pd.read_csv('../data/interim/cleaned_pbp_data.csv')
try:
    df = df.drop('Unnamed: 0',axis=1)
except:
    pass

df['Possession_Value'] = 0

for idx,row in df.iterrows():
    if row['FT_Made'] == 1:
        df.at[idx,'Possession_Value'] = 1
    if row['2P_Made'] == 1:
        df.at[idx,'Possession_Value'] = 2
    if row['3P_Made'] == 1:
        df.at[idx,'Possession_Value'] = 3
        
df.to_csv('../data/interim/cleaned_pbp_data.csv')
```
